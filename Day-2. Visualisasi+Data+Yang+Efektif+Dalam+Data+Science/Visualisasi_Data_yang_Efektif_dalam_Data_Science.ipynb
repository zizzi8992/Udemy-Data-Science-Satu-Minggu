```python
# Metode plot pada Pandas Series dan DataFrames hanyalah wrapper dari plt.plot()
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

## TUGAS #4. PLOT HISTOGRAM DENGAN MATPLOTLIB

```python
# Histogram mewakili data menggunakan batang (bar) dengan berbagai ketinggian
# Setiap bar mengelompokkan angka ke dalam rentang tertentu
# Bar yang lebih tinggi menunjukkan bahwa lebih banyak data berada dalam rentang spesifik tersebut

daily_return_df = pd.read_csv('crypto_daily_returns.csv')
daily_return_df
```

```python
mu = round(daily_return_df['BTC'].mean(), 2) # mean of distribution
sigma = round(daily_return_df['BTC'].std(), 2) # standard deviation of distribution
 
num_bins = 100

# Plot histogram dari return
plt.figure(figsize = (20, 15)) 

daily_return_df['BTC'].plot.hist(bins = num_bins, color = 'b', alpha = 0.7, figsize = (12, 6))

plt.grid()
plt.ylabel('Probability')
plt.title('Histogram: mu=' + str(mu) + ', sigma=' +str(sigma));
 
```

```python
values  = [10, 70, 10, 10]
colors  = ['g', 'r', 'y', 'b']
explode = [0, 0.2, 0, 0]
labels  = ['JP Morgan Chase (JPM)', 'APPLE (AAPL)', 'United Airlines (UAL)', 'P&G (PG)']

# Use matplotlib to plot a pie chart 
plt.figure(figsize = (10, 10))
plt.pie(values, colors = colors, labels = labels, explode = explode)
plt.title('STOCK PORTFOLIO')
plt.show()
```

```python
# Plot the histogram of the returns
plt.figure(figsize = (20, 15)); 

mu_UAL = round(daily_return_df['UAL'].mean(), 2) # mean of distribution
sigma_UAL = round(daily_return_df['UAL'].std(), 2) # standard deviation of distribution
 
    
mu_PG = round(daily_return_df['PG'].mean(), 2) # mean of distribution
sigma_PG = round(daily_return_df['PG'].std(), 2) # standard deviation of distribution
     
    
daily_return_df[['PG','UAL']].plot.hist(bins = 60, alpha = 0.7, figsize = (15, 8));
plt.grid();
plt.ylabel('Probability')


plt.title('Histogram: P&G mu=' + str(mu_PG) + ', P&G sigma=' +str(sigma_PG) + ', United Airlines mu=' +str(mu_UAL) + ', United Airlines sigma=' +str(sigma_UAL) );
```
