```python
#IMPORT PD, NUMPY, SEABORN DAN MATPOTLIB
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#imprt data daily prices
daily_prices_df = pd.read_csv('stocks_daily_prices.csv')
daily_prices_df.head()
daily_prices_df.tail()
```

```python
#plot multiple lineplot 4 saham dalam satu figure
daily_prices_df.plot(x = 'Date', y = ['AAPL', 'JPM', 'PG', 'UAL'], linewidth = 3, figsize = (14,6));
plt.ylabel('Price [$]')
plt.title('Daily Prices')
plt.legend(loc = 'upper left')
plt.grid()
```

```python
#plot harga daily_prices_df dalam subplot
daily_prices_df.plot(x = 'Date', y = ['AAPL', 'JPM', 'PG', 'UAL'], linewidth = 3, figsize = (14,6), subplots = True);
```

```python
#plot semua subplot dalam satu baris using matpotlib
plt.figure(figsize = (20,4))

plt.subplot(1, 4, 1) #barisnya 1 kolomnya 4 index 1
plt.plot(daily_prices_df['AAPL'], 'r--') #warna merah tipe -- garis putus2
plt.grid()

plt.subplot(1, 4, 2) #barisnya 1 kolomnya 4 index 2
plt.plot(daily_prices_df['JPM'], 'b.') #warna biru tipe .. garis titik2
plt.grid()

plt.subplot(1, 4, 3) #barisnya 1 kolomnya 4 index 3
plt.plot(daily_prices_df['PG'], 'g.') #warna hijau tipe .. garis titik2
plt.grid()

plt.subplot(1, 4, 4) #barisnya 1 kolomnya 4 index 4
plt.plot(daily_prices_df['UAL'], 'y--') #warna kuning tipe -- garis putus2
plt.grid()
```

```python
#EDA return harian
daily_returns_df = pd.read_csv('stocks_daily_returns.csv')
daily_returns_df.head()
daily_returns_df.tail()
```

```python
#scatterplot AAPL vs JPM (pt)
daily_returns_df.plot.scatter('AAPL', 'JPM', grid = True, figsize = (12, 6), color = 'hotpink', alpha = 0.5)
```

```python
#scatterplot AAPL vs JPM (sns)
plt.figure(figsize = (12,6))
plt.grid()
sns.scatterplot(x = 'AAPL', y = 'JPM', data = daily_returns_df, alpha = 0.5)
```

```python
#bullish 70% saham AAPL
%matplotlib inline
stocks_df = pd.DataFrame(data = { 'allocation %' : [70, 10, 10, 10]}, index = ['AAPL', 'JPM', 'PG', 'UAL'])
stocks_df
```

```python
from seaborn._core.properties import Color
#plot piechart using matpotlib
stocks_df.plot.pie( y = 'allocation %', figsize = (6,6), explode = [0.1, 0, 0, 0])
plt.title("STOCKS PORTOFOLIO PIE CHART")
```

```python
#plot histogram UAL vs PG using matpotlib
muUAL = round(daily_returns_df['UAL'].mean(), 2)
muPG = round(daily_returns_df['PG'].mean(), 2)
sigmaUAL = round(daily_returns_df['UAL'].std(), 2)
sigmaPG = round(daily_returns_df['PG'].std(), 2)
num_bins = 40
plt.figure(figsize = (20,15))
daily_returns_df[['UAL', 'PG']].plot.hist(bins = num_bins, color = ['r', 'b'], alpha = 0.5)
plt.grid()
plt.ylabel('Probability')
plt.title('Histogram: muUAL =' + str(muUAL) + ' dan sigmaUAL =' + str(sigmaUAL) + ' & Histogram: muPG =' + str(muPG) + ' dan sigmaPG =' + str(sigmaPG));
```

```python
#plot heatmap yang menunjukkan korelasi antara return harian saham using seaborn
plt.figure(figsize = (20, 10))
sns.heatmap(daily_returns_df.corr(), annot = True)
```

```python
#plot 3D
from mpl_toolkits.mplot3d import Axes3D
sns.set_style("whitegrid", {'axes.grid' : False})

fig = plt.figure(figsize=(15,15))

ax = fig.add_subplot(111, projection = '3d')

x = daily_returns_df['JPM'].tolist() #dibuat jadi list (array) datanya
y = daily_returns_df['AAPL'].tolist()
z = daily_returns_df['UAL'].tolist()


ax.scatter(x, y, z, c = 'r', marker='o', s = 10) # c=color, tanda/marker = o, s=besar bulatannya(marker float)
ax.set_xlabel('JPM')
ax.set_ylabel('AAPL')
ax.set_zlabel('UAL')

plt.show()
```
