```python
#buat one hot encoding untuk 'Status'
life_expectancy_df['Status'].unique()
```

```python
life_expectancy_df = pd.get_dummies(life_expectancy_df, columns = ['Status'])
life_expectancy_df
#sekarang kolom status telah di encoding menjadi developed = 0 dan developing=1
```

```python
#Bagi data menjadi 80% data latih dan 20% data testing
life_expectancy_df.isnull().sum()[np.where(life_expectancy_df.isnull().sum() != 0)[0]]
#melihat jumlah nilai null untuk kolom yg memiliki nilai null
#lalu hanya diambil dimna yg jumlahnya tidak sama dgn null
```

```python
#mengisi nilai null dgn nilai mean, karena sebagian besar nilainya kontinyu
life_expectancy_df = life_expectancy_df.apply(lambda x: x.fillna(x.mean()), axis = 0)
```

```python
#cek ulang nilai null
life_expectancy_df.isnull().sum()[np.where(life_expectancy_df.isnull().sum() != 0) [0]]
```

```python
#Buat data latih dan data uji
X = life_expectancy_df.drop(columns = ['Life expectancy '])
y = life_expectancy_df[['Life expectancy ']] #array dalam array supaya berda dlm kolom, kalo cuma ['Life expectancy'] aja langsung data diluar kolom munculnya
```

```python
X
```

```python
y.shape
```

```python
#konversi tipe data ke float32
X = np.array(X).astype('float32')
y = np.array(y).astype('float32')
```

```python
X #berisikan hanya var numerik
```

```python
#bagi data uji dan data latih
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2) #Bagi data menjadi 80% data latih dan 20% data testing
```

```python
from sklearn.metrics import mean_squared_error, accuracy_score
```

```python
#Latih model XG-boost
from xgboost.core import Objective
#latih model regresi xgboost
import xgboost as xgb
model = xgb.XGBRegressor(objective = 'reg:squarederror', learning_rate = 0.1, max_depth = 30, n_estimators = 100)
model.fit(X_train, y_train)
```

```python
model_accuracy = model.score(X_test, y_test)
model_accuracy
```

```python
#prediksi
y_predict = model.predict(X_test)
y_predict
```

```python
#Evaluasi model regresi yang telah dilatih, apa itu R2?
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from math import sqrt
```

```python
k = X_test.shape[1]
n = len(X_test)
RMSE = float(format(np.sqrt(mean_squared_error(y_test, y_predict)), '.3f'))
```

```python
MSE = mean_squared_error(y_test, y_predict)
MAE = mean_squared_error(y_test, y_predict)
r2 = r2_score(y_test, y_predict)
adj_r2 = 1-(1-r2)*(n-1)/(n-k-1)
```

Dengan mengganti nilai salah satu hyper parameter saja misal max_depth akan merubah hasil dari model yg dihasilkan, semakin tinggi nilai max_depth nya nilai modelnya juga meningkat

```python
#Plot prediksi dari model yang dilatih vs keluaran sebenarnya (ground-truth)
plt.plot(y_test, y_predict, "^", color = "r")
plt.title('Model Prediction vs True Value')
plt.xlabel('Model Prediction')
plt.ylabel('True Value')
plt.show()
```

```python
print('RMSE =',RMSE, '\nMSE =', MSE, '\nMAE =', MAE, '\nr2 =', r2, '\nadj_r2', adj_r2)
```
