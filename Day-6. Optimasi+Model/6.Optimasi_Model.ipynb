```python
#LATIHAN 4
#Pemisahan data latih dan uji
#kolom kategori
X_cat = bike_df[['season', 'yr', 'mnth', 'holiday', 'weekday', 'workingday', 'weathersit']]
X_cat
```

```python
#melakukan encoding
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder()
X_cat = onehotencoder.fit_transform(X_cat).toarray()
X_cat
```

```python
#sebelumnya hanya 7 kolom, setelah encoding menjadi 32 kolom
X_cat.shape
```

```python
#mengubah X_cat menjadi data frame
X_cat = pd.DataFrame(X_cat)
```

```python
X_numerical
```

```python
#mereset index num yg sebelumnya time index sekarang dikembalikan ke dalam kolom
X_numerical = X_numerical.reset_index()
X_numerical
```

```python
#menggabungkan X_cat dan num
X_all = pd.concat([X_cat, X_numerical], axis = 1)
X_all
```

```python
#karena data dteday tdk bisa difunakan untuk prediksi maka di drop
X_all = X_all.drop('dteday', axis = 1)
X_all
```

```python
#sekarang tersisa 36 kolom dimana 35 lainnya prediktor dan 1 kolom cnt sebagai target var
#menyimpan dalam array
X = X_all.iloc[:, :-1].values #[: (semua baris), :-1(kolom terakhir dikurangi atau ga dimasukin)]
y = X_all.iloc[:, -1:].values #[: (semua baris), -1:(kolom terakhir aja yg dimasukin)]
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X_train.shape #80% baris yaitu 584
```

```python
X_test.shape #20% baris yaitu 147
```

```python
#LATIHAN 5
#Algoritma xgb tanpa optimasi
!pip install xgboost
```

```python
#Latih model XG-boost
from xgboost.core import Objective
import xgboost as xgb
```

```python
model = xgb.XGBRegressor(objective = 'reg:squarederror', learning_rate = 1, max_depth = 20, n_estimators = 500)
```

```python
model.fit(X_train, y_train)
```

```python
#mengevaluasi model yang sudah dilatih pada data uji
result = model.score(X_test, y_test)
print('Accuracy: {}'.format(result))
```

```python
#melakukan prediksi data uji
y_predict = model.predict(X_test)
```

```python
y_test
```

```python
y_predict
```

```python
#menghitung metric
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from math import sqrt
k = X_test.shape[1]
n = len(X_test)
```

```python
k #kolom
```

```python
n #panjang x test
```

```python
RMSE = float(format(np.sqrt(mean_squared_error(y_test, y_predict)), '.3f'))
MSE = mean_squared_error(y_test, y_predict)
MAE = mean_absolute_error(y_test, y_predict)
r2 = r2_score(y_test, y_predict)
adj_r2 = 1-(1-r2)*(n-1)/(n-k-1)

print('RMSE =', RMSE, '\nMSE =', MSE, '\nMAE =', MAE, '\nR2 =', r2, '\nAdjusted R2 =', adj_r2)
```

```python
#LATIHAN 7
#randomsearch
from sklearn.model_selection import RandomizedSearchCV
grid = {'max_depth' : [2, 3, 5],
        'learning_rate' : [0.1, 0.5, 1],
        'n_estimators' : [100, 500, 700],
        'min_child_weight' : [1, 2, 3]}
```

```python
import xgboost as xgb
model = xgb.XGBRegressor(objective= 'reg:squarederror')
```

```python
random_cv = RandomizedSearchCV(estimator= model,
                               param_distributions= grid,
                               cv = 5,
                               n_iter = 50,
                               scoring = 'neg_mean_absolute_error',
                               verbose = 5,
                               return_train_score= True)
```

```python
random_cv.fit(X_train, y_train)
```
